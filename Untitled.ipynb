```python
import pandas as pd
from prophet import Prophet
from sklearn.metrics import mean_absolute_percentage_error
import matplotlib.pyplot as plt
import numpy as np
```

```python
# read in the data
df = pd.read_excel('/tmp/data.xlsx')
df["ds"]=df["Date"]
df["y"]=df["Gross Sales"]


# split into train and test set
# 121 training samples and 30 testing samples i.e. 80% 20% 
df_train = df[:121]
df_test = df[121:]


df_train.head()
```

```python
#show the data against time

plt.figure(figsize=(20,6))
plt.plot("Date","y","",data=df_train)
plt.plot("Date","y","",data=df_test)
plt.show()
```

```python
#compare to some baseline results: use the mean of the training data

y_mean = df_train["y"].mean()
y_mean_series = pd.Series([y_mean for i in range(31)])

baseline_score = mean_absolute_percentage_error(df_test["y"], y_mean_series)
print(f"the baseline mape is: {baseline_score}")
plt.plot("Date","y","",data=df_test)
plt.plot(df_test["Date"], y_mean_series)
plt.show()
```

```python
# use the prophet forecasting model - https://facebook.github.io/prophet/

m = Prophet()
m.fit(df_train)


forecast = m.predict(df_test)

results = pd.merge(forecast,df_test,on="ds")

score = mean_absolute_percentage_error(df_test["y"], results["yhat"])

print(f"the prophet mape is: {score}")

plt.plot("Date","y","",data=df_test)
plt.plot(results["Date"], results["yhat"])
plt.show()

#better, but still lots of room to improve
```

```python
# use the prophet forecasting model, this time with an extra regressor

m = Prophet(yearly_seasonality=False)
m.add_regressor('OM2W')
m.add_regressor('OM1W')

m.fit(df_train)


forecast = m.predict(df_test)

results = pd.merge(forecast,df_test,on="ds")

score = mean_absolute_percentage_error(df_test["y"], results["yhat"])

print(f"the improved prophet mape is: {score}")

plt.plot("Date","y","",data=df_test)
plt.plot(results["Date"], results["yhat"])
plt.show()

#a better result
```

```python
# use the prophet forecasting model, this time with an extra regressor

m = Prophet(yearly_seasonality=False)
m.add_regressor('OM2W')
m.add_regressor('OM1W')

m.fit(df_train[50:])


forecast = m.predict(df_test)

results = pd.merge(forecast,df_test,on="ds")

score = mean_absolute_percentage_error(df_test["y"], results["yhat"])

print(f"the improved prophet mape, ignoring the oldest training data, is: {score}")

plt.plot("Date","y","",data=df_test)
plt.plot(results["Date"], results["yhat"])
plt.show()

#a marginal improvement in mape by disregarding the oldest data
```
